```python
import pandas as pd
file_path ='/content/drive/MyDrive/python_loading/Customer Call List.xlsx'
```

```python
df = pd.read_excel(file_path)
#df
```

```python
# Xóa những hàng bị trùng
df = df.drop_duplicates()
#df
```

```python
# Xóa cột theo mong muốn cụ thể là cột 'Not_Useful_Column'
df = df.drop(columns='Not_Useful_Column')
df
```

```python
# Chuẩn hóa cột Last_name
df["Last_Name"]= df["Last_Name"].str.strip("/_.")
#df
```

```python
# Chuẩn hóa cột Phone_Number
df['Phone_Number'] = df['Phone_Number'].str.replace('[^0-9]','')
# [^a-zA-Z0-9]: Là một biểu thức chính quy (regular expression):
# ^: Đại diện cho việc phủ định hoặc loại trừ những ký tự sau nó.
# a-z: Tất cả các chữ cái viết thường.
# A-Z: Tất cả các chữ cái viết hoa.
# 0-9: Tất cả các số từ 0 đến 9.
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))# chuyển sang dạng str
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--','')
df['Phone_Number'] = df['Phone_Number'].str.replace('--','')
#df
```

```python
# chuẩn hóa cột Address
# tách thành 3 cột mới là
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', expand=True)
                                                                          # expand=True: Kết quả tách chuỗi sẽ được trả về dưới dạng DataFrame.
df["Address"] = df["Address"].str.replace("N/a","")
#df
```

```python
# Chuẩn hóa 2 cột Paying Customer, Do_Not_Contact

# đây là cách thứ hai có thể replace đồng thời
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y').str.replace('No', 'N').str.replace('N/a','')

df['Do_Not_Contact'] = df['Do_Not_Contact'].apply(lambda x: str(x))
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y').str.replace('No', 'N').str.replace('nan','')

df
```

```python
df['Street_Address'] = df['Street_Address'].str.replace('N/a','')

# cách 2 thay được toàn bộ mà k cần phải thay từng ký tự Na,....
df=df.fillna('')
df
```
